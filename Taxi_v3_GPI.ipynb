```
import gym
import random
```

```
# Initialize the environment 

env = gym.make("Taxi-v3")
```

```
# Understand the environment 

observation_space = env.observation_space
observation_space
```

```
action_space = env.action_space
action_space
```

```
num_states = observation_space.n
num_actions = action_space.n
```

```
import numpy as np
policy = np.random.randint(6, size=num_states)
policy
```

```
def generate_episode(policy, render=False):
  
  episode = [] 
  s0 = env.reset()
  if render:
    env.render()
  for i in range(50):
    action = policy[s0]
    next_state, reward, done, debug = env.step(action) 
    episode.append([s0, action, reward, next_state])

    s0 = next_state 

    if render:
      env.render()
    if done:
      break

  return np.array(episode) 
```

```
value_function = np.zeros((num_states))
```

```


def q_learning_policy_evaluation(policy, num_states, num_actions, env, q_val_func, epsilon):
  max_steps=50
  lr=0.1
  num_episodes=100

  
  #rewards_all_episodes = []
  for i in range(num_episodes):
    #reward=0;

    st=env.reset()
    for j in range(max_steps):


      
      exploration_rate_threshold = random.uniform(0, 1)
      if exploration_rate_threshold > epsilon:
        at = np.argmax(q_val_func[st,:]) 
      else:
        at = env.action_space.sample()

      st1, rt, done, debug =env.step(at)
      old_value = q_val_func[st, at]

      #temporal difference formula
      new_value = (1-lr)*old_value + lr*(rt + 0.75*np.max(q_val_func[st1,:]))

      q_val_func[st,at] = new_value
      st=st1
     #reward+=rt
      if done:
        break

    #rewards_all_episodes.append(reward)


  return q_val_func
```

```
def policy_improvement(policy, q_val_func,epsilon, env):
  improved_policy = np.copy(policy)

  for state in range(num_states):
    

    improved_policy[state]=np.argmax(q_val_func[state, :])
    
  return improved_policy
```

```
def q_learning(num_actions, num_states, env):
  num_iterations= 100
  policy = np.random.randint(num_actions, size=num_states)
  epsilon =1
  epsilon_decay = 0.01
  q_val_func = np.zeros((num_states,num_actions))
  for i in range(num_iterations):
    q_val_func = q_learning_policy_evaluation(policy, num_states, num_actions, env,q_val_func, epsilon)
    policy = policy_improvement(policy,q_val_func, epsilon, env) 
    epsilon = epsilon -epsilon_decay
    

  return policy

```

```
#train
optimal_policy = q_learning(num_actions,num_states,env)
#test
episode = generate_episode(optimal_policy,render= True)
```
